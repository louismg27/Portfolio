## Justificar borrado de columnas
Ahora, en un caso, cuando las variables independientes estÃ¡n correlacionadas (esta condiciÃ³n es multicolinealidad), indica que los cambios en una variable tambiÃ©n causarÃ­an cambios en otra variable. 
Cuanto mÃ¡s fuerte sea esta correlaciÃ³n, mayor serÃ¡ la dificultad de cambiar una variable sin cambiar otra. Esto hace que sea difÃ­cil para el modelo de regresiÃ³n estimar la relaciÃ³n entre cada variable independiente y la variable dependiente de forma independiente, reduciendo asÃ­ el rendimiento de su modelo de regresiÃ³n.

## STANDARD SCALER
La idea detrÃ¡s de StandardScaler es que transformarÃ¡ sus datos de modo que su distribuciÃ³n tenga un valor medio 0 y una desviaciÃ³n estÃ¡ndar de 1.
En el caso de datos multivariados, esto se hace en funciÃ³n de las caracterÃ­sticas (en otras palabras, independientemente para cada columna de datos).
Dada la distribuciÃ³n de los datos, cada valor en el conjunto de datos tendrÃ¡ el valor medio restado, y luego dividido por la desviaciÃ³n estÃ¡ndar de todo el conjunto de datos (o caracterÃ­stica en el caso multivariante).

```python
## En lenguaje R serÃ­a 

# dataNumeric = data %>% select(is.character)
# scaler = scale(dataNumeric)

#Reescalar columnas numÃ©ricas

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

#Columnas numÃ©ricas
conFeature = data.columns[data.dtypes != object]
col = conFeature.tolist()

#Aplicamos Standard Scaler 
dataNumeric = data[col]
sc = StandardScaler()
scaler = sc.fit(dataNumeric)
scaler.transform(dataNumeric)
```
