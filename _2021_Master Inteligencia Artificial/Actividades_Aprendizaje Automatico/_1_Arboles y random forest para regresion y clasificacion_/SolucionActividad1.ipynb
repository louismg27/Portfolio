## Justificar borrado de columnas
Ahora, en un caso, cuando las variables independientes están correlacionadas (esta condición es multicolinealidad), indica que los cambios en una variable también causarían cambios en otra variable. 
Cuanto más fuerte sea esta correlación, mayor será la dificultad de cambiar una variable sin cambiar otra. Esto hace que sea difícil para el modelo de regresión estimar la relación entre cada variable independiente y la variable dependiente de forma independiente, reduciendo así el rendimiento de su modelo de regresión.

## STANDARD SCALER
La idea detrás de StandardScaler es que transformará sus datos de modo que su distribución tenga un valor medio 0 y una desviación estándar de 1.
En el caso de datos multivariados, esto se hace en función de las características (en otras palabras, independientemente para cada columna de datos).
Dada la distribución de los datos, cada valor en el conjunto de datos tendrá el valor medio restado, y luego dividido por la desviación estándar de todo el conjunto de datos (o característica en el caso multivariante).

```python
## En lenguaje R sería 

# dataNumeric = data %>% select(is.character)
# scaler = scale(dataNumeric)

#Reescalar columnas numéricas

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

#Columnas numéricas
conFeature = data.columns[data.dtypes != object]
col = conFeature.tolist()

#Aplicamos Standard Scaler 
dataNumeric = data[col]
sc = StandardScaler()
scaler = sc.fit(dataNumeric)
scaler.transform(dataNumeric)
```

```python
#Separar datos de entrenamiento y de test

xTrain,xTest,yTrain,yTest = train_test_split(newData,data['SalePrice'],test_size=0.2)
```
